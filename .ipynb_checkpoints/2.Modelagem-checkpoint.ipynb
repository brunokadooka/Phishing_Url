## 2. Modelagem

```python
# Ferramentas que vamos utilizar na modelagem

# Tecnicas de sampling
from imblearn.under_sampling import NearMiss, RandomUnderSampler
from imblearn.over_sampling import SMOTE, RandomOverSampler

# Pipeline do pacote imblearn para realização de tecnicas de Samplig (nossos dados são desbalanceados)
from imblearn.pipeline import Pipeline as ImbalancedPipeline

# Scalers
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

# Seleção das melhores features
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif

# Validação Cruzada
from sklearn.model_selection import cross_validate, StratifiedKFold, cross_val_score, cross_val_predict

# Busca de Hiperparametros Bayesiana
from skopt import gp_minimize

# Metricas para avaliação
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix
```

```python
# Funcao para imprimir os resultados e o intervalo
def imprime_resultados(results, weigth = 2):
    media = results['test_score'].mean()
    desvio_padrao = results['test_score'].std()
    print("Accuracy mean: %.2f%%" % (media * 100))
    print("Accuracy std: %.4f" % (desvio_padrao))
    print("Accuracy Interval Cross Validation: [%.2f%%, %.2f%%]" % ((media - weigth * desvio_padrao)*100, (media + weigth * desvio_padrao)*100))
```

### 2.1 Modelos Dummy

Estes modelos dummy servem para dar uma baseline para nosso algoritmo. Claro, a baseline geral, visto que este modelo atende como chute ou alguma previsão aleatória, o que se espera é que outros modelos performam melhores, modelos que não consigam performar melhores que modelos dummy, pode-se classificar no sentido de **underfitting**.

Além disso os modelos dummy tem duas formas:

- most-frequent: Este modelo ele prevê apenas um valor de classe, ou seja, é o valor mais frequente, no nosso dataset conforme vimos na seção 1, a classe mais frequente é a 0. Então este modelo ele vai prever os datasets sempre chutando o valor de classe 0 (Ou seja, ele só vai dizer que o site é phishing). Visto que o dataset tem 67% de valores, logo, a acuracia vai ser de 67%;

- stratified: Ele leva em conta a probabilidade de determinado registro ser da classe 0 ou da classe 1, para isso, ele cria uma coluna one-hot. Ou seja, este algoritmo vai calcular a probabilidade de pertencer a classe 0 ou a classe 1, aquela que tiver a maior probabilidade o algoritmo vai chutar nele. Este é um modelo que mais "tenta prever" nos modelos aleatório, é quase com um chute mas calculado.

```python
# importacao dos modelos dummy
from sklearn.dummy import DummyClassifier
```

```python
SEED = 301
np.random.seed(SEED)

# modelo dummy most frequent
dummy_most_frequent = DummyClassifier(strategy='most_frequent')
dummy_most_frequent.fit(X, y)
y_pred = dummy_most_frequent.predict(X_test)
```

```python
# Plotando a matriz de confusao
sns.heatmap(confusion_matrix(y_test, y_pred), cmap='OrRd', annot=True, fmt='2.0f', 
              xticklabels=['Phishing', 'Normal'], yticklabels=['Phishing', 'Normal'])

acc = accuracy_score(y_test, y_pred).round(2)
plt.title('Dummy Classifier Most Frequent - Acc: %.2f%%' % (acc*100))
plt.ylabel('Real')
plt.xlabel('Previsto')
plt.show()
```

### 2.2 Modelo KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

#### 2.2.1 Sem Hiperparametros

```python
SEED = 300
np.random.seed(SEED)

# Separando o modelo
model_knn =  KNeighborsClassifier()

# Treinando com a base de treinamento
model_knn.fit(X, y)

# Predicoes
y_pred_model_one = model_knn.predict(X_test)
```

```python
# Verificação de overfitting

# Verificando a acuracia de Treino e Test
acc_train = model_knn.score(X, y)
acc_test = accuracy_score(y_test, y_pred_model_one)
auc_roc = roc_auc_score(y_test, y_pred_model_one)

# Visualizando as acuracias
print("Acuracia de Treino: %.2f%%" % (acc_train * 100))
print("Acuracia de Teste: %.2f%%" % (acc_test * 100))
print("AUC-ROC: %.2f%%" % (auc_roc * 100))
```

#### 2.2.2 Com Hiperparametros

```python
# Definindo qual sampler vamos utilizar
def return_sampling (type_sampling, SEED):

    if type_sampling == 'NearMiss': 
        sampling = NearMiss()
    elif type_sampling == 'SMOTE':
        sampling = SMOTE(random_state=SEED)
    elif type_sampling == 'RandomUnder':
        sampling = RandomUnderSampler(random_state=SEED)
    elif type_sampling == 'RandomOver':
        sampling = RandomOverSampler(random_state=SEED)
    else:
        sampling = None
        
    return sampling


# Definindo qual scaling vamos utilizar
def return_scaling (type_scaler):
    if type_scaler == 'StandardScaler':
        scaler = StandardScaler()
    elif type_scaler == 'MinMaxScaler':
        scaler = MinMaxScaler()
    elif type_scaler == 'RobustScaler':
        scaler = RobustScaler()
    else:
        scaler = None
        
    return scaler


# Definindo qual score_func do SelectKBest vamos utilizar
def return_score_func (type_score_func):
    # Definindo Funcao de score do SelectKBest
    if type_score_func == 'f_classif':
        score_func = f_classif
    elif type_score_func == 'mutual_info_classif':
        score_func = mutual_info_classif
        
    return score_func
```

```python
# Funcao para buscar melhor modelo
def treinar_modelo_knn (params):
    try:
        # Definindo o random state
        SEED = 300
        
        # Setando os parametros
        n_neighbors = params[0]
        p = params[1]
        k = params[2]
        type_sampling = params[3]
        type_score_func = params[4]
        type_scaler = params[5]
        type_scaler_2 = params[6]
        
        # Definindo o sampling
        sampling = return_sampling (type_sampling, SEED)
        
        # Definindo os scalings
        scaler = return_scaling (type_scaler)
        scaler2 = return_scaling (type_scaler_2)
            
        # Definindo a funcao do SelectKBest
        score_func = return_score_func (type_score_func)
                    
        # Verificando os parametros
        print(params)
        
        # Testando os melhores parametros no pipeline do knn
        knn_pipeline_best = ImbalancedPipeline([
            ('sampler', sampling),
            ('scaler', scaler),
            ('scaler2', scaler2),
            ('selector', SelectKBest(k=k, score_func=score_func)),
            ('classifier', KNeighborsClassifier(n_neighbors=n_neighbors, p=p))
        ])
        
        # Utilizando uma validação cruzada estratificada - por causa do modelo desbalanceado
        cv = StratifiedKFold(n_splits=5, shuffle = True, random_state=SEED)
                
        # Realizando a validação cruzada dos modelos: procurando o melhor modelo com base na curva auc-roc
        y_probas = cross_val_predict(knn_pipeline_best, X, y, cv=cv, method='predict_proba')
        
        # Calculo das curvas auc-roc
        roc_auc = roc_auc_score(y, y_probas[:, 1])

        return -roc_auc
        
    except:
        return 1000.0
```

```python
# Espaço para testar
space = [
    (3, 100, 'log-uniform'), # n_neighbors
    (1, 2), # parametro p
    (1, 7), # parametro k
    ([None, 'NearMiss', 'SMOTE', 'RandomUnder', 'RandomOver']), # type_sampling
    (['f_classif', 'mutual_info_classif']), # type_score_func
    ([None, 'StandardScaler', 'MinMaxScaler', 'RobustScaler']), # type_scaler
    ([None, 'StandardScaler', 'MinMaxScaler', 'RobustScaler']), # type_scaler_2
]
```

```python
%%time
# Procurar melhor resultado --> score_func como f_classif
resultado_gp = gp_minimize(treinar_modelo_knn, space, verbose=1, n_calls=100, n_random_starts=60, n_jobs=-1, random_state=0)
```

```python
# Verificando melhores resultados
best_params = resultado_gp.x

# Visualizando
best_params
```

```python
# Reconfigurar as mesmas configurações

# Ponto de aleatoriedade
SEED = 300
np.random.seed(SEED)

# Selecionando o melhor modelo com base nos parametros
n_neighbors = best_params[0]
p = best_params[1]
k = best_params[2]
sampling = return_sampling (best_params[3], SEED)
score_func = return_score_func (best_params[4])
scaler1 = return_scaling (best_params[5])
scaler2 = return_scaling (best_params[6])


# Repassando os parametros para o pipeline do knn
knn_pipeline_best = ImbalancedPipeline([
    ('sampler', sampling),
    ('scaler', scaler1),
    ('scaler2', scaler2),
    ('selector', SelectKBest(k=k, score_func=score_func)),
    ('classifier', KNeighborsClassifier(n_neighbors=n_neighbors, p=p))
])

# Utilizando a mesma validação cruzada
cv = StratifiedKFold(n_splits=5, shuffle = True, random_state=SEED)

# Verificando se estamos com o mesmo modelo
y_probas = cross_val_predict(knn_pipeline_best, X, y, cv=cv, method='predict_proba')

# Calculo das curvas auc-roc
roc_auc = roc_auc_score(y, y_probas[:, 1])
        
print(f'Score AUC-ROC: {round(roc_auc, 4) * 100}%')
```

```python
# Treinando o melhor modelo
# Treinando com a base de treinamento
knn_pipeline_best.fit(X, y)

# Predicoes
y_pred_model_two = knn_pipeline_best.predict(X_test)


# Verificação de overfitting

# Verificando a acuracia de Treino e Test
acc_train = knn_pipeline_best.score(X, y)
acc_test = accuracy_score(y_test, y_pred_model_two)

# Visualizando as acuracias
print("Acuracia de Treino: %.2f%%" % (acc_train * 100))
print("Acuracia de Teste: %.2f%%" % (acc_test * 100))
print("AUC-ROC: %.2f%%" % (roc_auc * 100))
```

```python
print("RELATÓRIO DE CLASSIFICAÇÃO:\n\n")
print("Modelo Sem Hiperparametros:\n")
print(classification_report(y_test, y_pred_model_one))
print("--------------------------------------------------------\n")
print("Modelo Com Hiperparametros:\n")
print(classification_report(y_test, y_pred_model_two))
```

```python
# Plotando a matriz de confusao - Sem Hiperparametros
fig, axs = plt.subplots(1, 2, figsize = (10, 5))

# Plotando a matriz de confusao
sns.heatmap(confusion_matrix(y_test, y_pred_model_one), cmap='OrRd', annot=True, fmt='2.0f', 
              xticklabels=['Phishing', 'Normal'], yticklabels=['Phishing', 'Normal'], ax=axs[0], cbar=False)

acc = accuracy_score(y_test, y_pred_model_one).round(4)
axs[0].set_title('KNN Sem Hiperparametros - Acc: %.2f%%' % (acc*100), fontsize=10)
axs[0].set_ylabel('Real')
axs[0].set_xlabel('Previsto')

# Plotando a matriz de confusao - Com Hiperparametros
sns.heatmap(confusion_matrix(y_test, y_pred_model_two), cmap='OrRd', annot=True, fmt='2.0f', 
              xticklabels=['Phishing', 'Normal'], yticklabels=['Phishing', 'Normal'], ax=axs[1], cbar=False)

acc = accuracy_score(y_test, y_pred_model_two).round(4)
axs[1].set_title('KNN Com Hiperparametros - Acc: %.2f%%' % (acc*100), fontsize=10)
axs[1].set_ylabel('Real')
axs[1].set_xlabel('Previsto')

# Titulo da figura
fig.suptitle('Figura 1: Análise do Modelo KNN Classifier', fontsize=12)

plt.tight_layout()

plt.show()
```

**Sobre a figura**

Olhando o relatório e a figura 1, a primeira vista notamos que o *`modelo sem Hiperparametros`* é melhor em relação ao *`modelo com Hiperparametros`* tendo como base a medida de acurácia. Mas este medida engana.

Primeiro a medida que usamos para pegar o melhor modelo é curva auc-roc e se olharmos na sessão 2.2.1 o *`modelo sem Hiperparametros`* apresentou um auc-roc de 72.90%, enquanto o *`modelo com Hiperparametros`* teve um auc-roc de 81,28%. Mas como podemos perceber que o *`modelo com Hiperparametros`* se saiu melhor?

Bom vamos olhar agora para o relatório de classificação, as medidas de *precision*, *recall* e *f1-score*. O *`modelo sem Hiperparametros`* tem medidas muito boas para a classe 0, sendo ela majoritária, isso de fato acontece porque neste modelo não houve tratamento no desbalanceamento dos dados, portanto, este modelo está enviesado em prever melhor a classe 0 (Sites Phishings), entretanto, ao tentar prever a classe 1 (Sites Normais) o modelo já falha.

Já o *`modelo com Hiperparametros`* ele tenta equilibrar melhor as sua previsões com respeito as duas classes, ele acertou mais com relação em prever sites "Normais", sendo assim ele não está enviesado pela classe majoritária. Entretanto, visto que o *`modelo sem Hiperparametros`* é a baseline para o *`modelo com Hiperparametros`* este modelo não se saiu muito bem.

**Ps: O KNN superou o modelo Dummy, lembrando que o modelo Dummy é a base de comparação geral para os modelos, mas espera-se que o modelo com ajuste de hiperparametro se sobressaia sobre os modelos sem hiperparametros.**

### 2.3 Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

#### 2.3.1 Sem Hiperparametros

```python
SEED = 300
np.random.seed(SEED)

# Separando o modelo
model_tree =  DecisionTreeClassifier(random_state=SEED)

# Treinando com a base de treinamento
model_tree.fit(X, y)

# Predicoes
y_pred_model_one = model_tree.predict(X_test)
```

```python
# Verificação de overfitting

# Verificando a acuracia de Treino e Test
acc_train = model_tree.score(X, y)
acc_test = accuracy_score(y_test, y_pred_model_one)
auc_roc = roc_auc_score(y_test, y_pred_model_one)

# Visualizando as acuracias
print("Acuracia de Treino: %.2f%%" % (acc_train * 100))
print("Acuracia de Teste: %.2f%%" % (acc_test * 100))
print("AUC-ROC: %.2f%%" % (auc_roc * 100))
```

#### 2.3.2 Com Hiperparametros

```python
# Funcao para buscar melhor modelo
def treinar_modelo_tree (params):
    try:
        # Definindo o random state
        SEED = 300
        
        # Setando os parametros
        criterion = params[0]
        max_depth = params[1]
        min_samples_leaf = params[2]
        max_features = params[3]
        class_weight = params[4]
        type_sampling = params[5]
        
        
        # Definindo o sampling
        sampling = return_sampling (type_sampling, SEED)
        
        # Caso o max_depth venha com o valor 0, ele irá atribuir como None valor default
        max_depth = None if max_depth == 0 else max_depth
        
        # Verificando os parametros
        print(params)
        
        # Testando os melhores parametros no pipeline do knn
        tree_pipeline_best = ImbalancedPipeline([
            ('sampler', sampling),
            ('classifier', DecisionTreeClassifier(criterion=criterion,max_depth=max_depth, min_samples_leaf=min_samples_leaf,
                                                max_features=max_features, class_weight=class_weight, random_state=SEED))
        ])
        
        # Utilizando uma validação cruzada estratificada - por causa do modelo desbalanceado
        cv = StratifiedKFold(n_splits=5, shuffle = True, random_state=SEED)
                
        # Realizando a validação cruzada dos modelos: procurando o melhor modelo com base na curva auc-roc
        y_probas = cross_val_predict(tree_pipeline_best, X, y, cv=cv, method='predict_proba')
        
        # Calculo das curvas auc-roc
        roc_auc = roc_auc_score(y, y_probas[:, 1])

        return -roc_auc
        
    except:
        return 1000.0
```

```python
# Espaço para testar
space = [
    (['gini', 'entropy']), # criterion
    (0, 300), # max_depth
    (1, 100), # min_samples_leaf
    ([None, 'log2', 'sqrt']), # max_features
    ([None, 'balanced']), # class_weight
    (['NearMiss', 'SMOTE', 'RandomUnder', 'RandomOver']), # type_sampling = params[4]
]
```

```python
%%time
# Procurar melhor resultado --> score_func como f_classif
resultado_gp = gp_minimize(treinar_modelo_tree, space, verbose=1, n_calls=100, n_random_starts=60, n_jobs=-1, random_state=0)
```

```python
# Verificando melhores resultados
best_params = resultado_gp.x

# Visualizando
best_params
```

```python
# Reconfigurar as mesmas configurações

# Ponto de aleatoriedade
SEED = 300
np.random.seed(SEED)

# Selecionando o melhor modelo com base nos parametros
criterion = best_params[0]
max_depth = best_params[1]
min_samples_leaf = best_params[2]
max_features = best_params[3]
class_weight = best_params[4]
sampling = return_sampling (best_params[5], SEED)


# Caso o max_depth venha com o valor 0, ele irá atribuir como None valor default
max_depth = None if max_depth == 0 else max_depth

# Repassando os parametros para o pipeline do knn
tree_pipeline_best = ImbalancedPipeline([
    ('sampler', sampling),
    ('classifier', DecisionTreeClassifier(criterion=criterion,max_depth=max_depth, min_samples_leaf=min_samples_leaf,
                                                max_features=max_features, class_weight=class_weight, random_state=SEED))
])

# Utilizando a mesma validação cruzada
cv = StratifiedKFold(n_splits=5, shuffle = True, random_state=SEED)

# Verificando se estamos com o mesmo modelo
y_probas = cross_val_predict(tree_pipeline_best, X, y, cv=cv, method='predict_proba')

# Calculo das curvas auc-roc
roc_auc = roc_auc_score(y, y_probas[:, 1])
        
print(f'Score AUC-ROC: {round(roc_auc, 4) * 100}%')
```

```python
# Treinando o melhor modelo
# Treinando com a base de treinamento
tree_pipeline_best.fit(X, y)

# Predicoes
y_pred_model_two = tree_pipeline_best.predict(X_test)


# Verificação de overfitting

# Verificando a acuracia de Treino e Test
acc_train = tree_pipeline_best.score(X, y)
acc_test = accuracy_score(y_test, y_pred_model_two)

# Visualizando as acuracias
print("Acuracia de Treino: %.2f%%" % (acc_train * 100))
print("Acuracia de Teste: %.2f%%" % (acc_test * 100))
print("AUC-ROC: %.2f%%" % (roc_auc * 100))
```

```python
print("RELATÓRIO DE CLASSIFICAÇÃO:\n\n")
print("Modelo Sem Hiperparametros:\n")
print(classification_report(y_test, y_pred_model_one))
print("--------------------------------------------------------\n")
print("Modelo Com Hiperparametros:\n")
print(classification_report(y_test, y_pred_model_two))
```

```python
# Plotando a matriz de confusao - Sem Hiperparametros
fig, axs = plt.subplots(1, 2, figsize = (10, 5))

# Plotando a matriz de confusao
sns.heatmap(confusion_matrix(y_test, y_pred_model_one), cmap='OrRd', annot=True, fmt='2.0f', 
              xticklabels=['Phishing', 'Normal'], yticklabels=['Phishing', 'Normal'], ax=axs[0], cbar=False)

acc = accuracy_score(y_test, y_pred_model_one).round(4)
axs[0].set_title('Decision Tree Sem Hiperparametros - Acc: %.2f%%' % (acc*100), fontsize=10)
axs[0].set_ylabel('Real')
axs[0].set_xlabel('Previsto')

# Plotando a matriz de confusao - Com Hiperparametros
sns.heatmap(confusion_matrix(y_test, y_pred_model_two), cmap='OrRd', annot=True, fmt='2.0f', 
              xticklabels=['Phishing', 'Normal'], yticklabels=['Phishing', 'Normal'], ax=axs[1], cbar=False)

acc = accuracy_score(y_test, y_pred_model_two).round(4)
axs[1].set_title('Decision Tree Com Hiperparametros - Acc: %.2f%%' % (acc*100), fontsize=10)
axs[1].set_ylabel('Real')
axs[1].set_xlabel('Previsto')

# Titulo da figura
fig.suptitle('Figura 2: Análise do Modelo Decision Tree Classifier', fontsize=12)

plt.tight_layout()

plt.show()
```

Primeiro a medida que usamos para pegar o melhor modelo é curva auc-roc e se olharmos na sessão 2.3.1 o *`modelo sem Hiperparametros`* apresentou um auc-roc de 80.53%, enquanto o *`modelo com Hiperparametros`* teve um auc-roc de 80.89%. Isto é os modelos em si são parecidos, mas há um problema que ocorre em *`modelo sem Hiperparametros`*.

Já o *`modelo com Hiperparametros`* notamos que ele não tem este problema, visto que a diferença entre as métricas de treino e teste são razoalmente equiparaveis. Portanto, este modelo de arvore realmente aprendeu com os dados, porém se verificarmos as métricas, o relatório e a matriz de confusão, este modelo está igual ao KNN, ele não se saiu bem.

**Ps: O `modelo com Hiperparametros` neste caso seria a melhor escolha, visto que, ele não sofre de overfitting. Portanto, este modelo se saiu melhor que o KNN e a baseline Dummy**

### 2.4 Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

#### 2.4.1 Sem Hiperparametros

```python
SEED = 300
np.random.seed(SEED)

# Separando o modelo
model_rf = RandomForestClassifier(random_state=SEED)

# Treinando com a base de treinamento
model_rf.fit(X, y)

# Predicoes
y_pred_model_one = model_rf.predict(X_test)
```

```python
# Verificação de overfitting

# Verificando a acuracia de Treino e Test
acc_train = model_rf.score(X, y)
acc_test = accuracy_score(y_test, y_pred_model_one)
auc_roc = roc_auc_score(y_test, y_pred_model_one)

# Visualizando as acuracias
print("Acuracia de Treino: %.2f%%" % (acc_train * 100))
print("Acuracia de Teste: %.2f%%" % (acc_test * 100))
print("AUC-ROC: %.2f%%" % (auc_roc * 100))
```

#### 2.4.2 Com Hiperparametros

```python
# Funcao para buscar melhor modelo
def treinar_modelo_rf (params):
    try:
        # Definindo o random state
        SEED = 300
        
        # Setando os parametros
        n_estimators = params[0]
        criterion = params[1]
        max_depth = params[2]
        min_samples_leaf = params[3]
        max_features = params[4]
        class_weight = params[5]
        type_sampling = params[6]
        
        
        # Definindo o sampling
        sampling = return_sampling (type_sampling, SEED)
        
        # Caso o max_depth venha com o valor 0, ele irá atribuir como None valor default
        max_depth = None if max_depth == 0 else max_depth
        
        # Verificando os parametros
        print(params)
        
        # Testando os melhores parametros no pipeline do knn
        rf_pipeline_best = ImbalancedPipeline([
            ('sampler', sampling),
            ('classifier', RandomForestClassifier(n_estimators=n_estimators, criterion=criterion, max_depth=max_depth,
                                                  min_samples_leaf=min_samples_leaf, max_features=max_features, 
                                                  class_weight=class_weight, random_state=SEED, n_jobs=-1))
        ])
        
        # Utilizando uma validação cruzada estratificada - por causa do modelo desbalanceado
        cv = StratifiedKFold(n_splits=5, shuffle = True, random_state=SEED)
                
        # Realizando a validação cruzada dos modelos: procurando o melhor modelo com base na curva auc-roc
        y_probas = cross_val_predict(rf_pipeline_best, X, y, cv=cv, method='predict_proba')
        
        # Calculo das curvas auc-roc
        roc_auc = roc_auc_score(y, y_probas[:, 1])

        return -roc_auc
        
    except:
        return 1000.0
```

```python
# Espaço para testar
space = [
    (100, 2000), # n_estimators
    (['gini', 'entropy']), # criterion
    (0, 300), # max_depth
    (1, 100), # min_samples_leaf
    ([None, 'log2', 'sqrt']), # max_features
    ([None, 'balanced']), # class_weight
    (['NearMiss', 'SMOTE', 'RandomUnder', 'RandomOver']), # type_sampling
]
```

```python
%%time
# Procurar melhor resultado --> score_func como f_classif
resultado_gp = gp_minimize(treinar_modelo_rf, space, verbose=1, n_calls=100, n_random_starts=60, n_jobs=-1, random_state=0)
```

```python
# Verificando melhores resultados
best_params = resultado_gp.x

# Visualizando
best_params
```

```python
# Reconfigurar as mesmas configurações

# Ponto de aleatoriedade
SEED = 300
np.random.seed(SEED)

# Selecionando o melhor modelo com base nos parametros
n_estimators = best_params[0]
criterion = best_params[1]
max_depth = best_params[2]
min_samples_leaf = best_params[3]
max_features = best_params[4]
class_weight = best_params[5]
sampling = return_sampling (best_params[6], SEED)


# Caso o max_depth venha com o valor 0, ele irá atribuir como None valor default
max_depth = None if max_depth == 0 else max_depth

# Repassando os parametros para o pipeline do knn
rf_pipeline_best = ImbalancedPipeline([
    ('sampler', sampling),
    ('classifier', RandomForestClassifier(n_estimators=n_estimators, criterion=criterion, max_depth=max_depth,
                                            min_samples_leaf=min_samples_leaf, max_features=max_features, 
                                            class_weight=class_weight, random_state=SEED, n_jobs=-1))
])

# Utilizando a mesma validação cruzada
cv = StratifiedKFold(n_splits=5, shuffle = True, random_state=SEED)

# Verificando se estamos com o mesmo modelo
y_probas = cross_val_predict(rf_pipeline_best, X, y, cv=cv, method='predict_proba')

# Calculo das curvas auc-roc
roc_auc = roc_auc_score(y, y_probas[:, 1])
        
print(f'Score AUC-ROC: {round(roc_auc, 4) * 100}%')
```

```python
# Treinando o melhor modelo
# Treinando com a base de treinamento
rf_pipeline_best.fit(X, y)

# Predicoes
y_pred_model_two = rf_pipeline_best.predict(X_test)


# Verificação de overfitting

# Verificando a acuracia de Treino e Test
acc_train = rf_pipeline_best.score(X, y)
acc_test = accuracy_score(y_test, y_pred_model_two)

# Visualizando as acuracias
print("Acuracia de Treino: %.2f%%" % (acc_train * 100))
print("Acuracia de Teste: %.2f%%" % (acc_test * 100))
print("AUC-ROC: %.2f%%" % (roc_auc * 100))
```

```python
print("RELATÓRIO DE CLASSIFICAÇÃO:\n\n")
print("Modelo Sem Hiperparametros:\n")
print(classification_report(y_test, y_pred_model_one))
print("--------------------------------------------------------\n")
print("Modelo Com Hiperparametros:\n")
print(classification_report(y_test, y_pred_model_two))
```

Primeiro, como sempre comentamos, a medida que usamos para pegar o melhor modelo é curva auc-roc, na sessão 2.4.2 o *`modelo sem Hiperparametros`* apresentou um auc-roc de 79.65%, enquanto o *`modelo com Hiperparametros`* teve um auc-roc de 83.97%. Entao vamos entender porque o *`modelo com Hiperparametros`* se saiu melhor agora.

Então como já citado o *`modelo sem Hiperparametros`* sofre novamente com **overfitting**. Acuracia de Treino em 94.92% e a de teste em 83.06%, esta diferença indica a presença de overfitting; Alguns dizem que o modelo Random Forest não *"overfitta"*, mas segundo Mario Filho, a verdade é que conforme aumentamos o numero de arvores, mais o modelo vai ficando estavel, ou evitando o overfitting (Esta é a diferença entre Random Forest para os modelos de Boost -- fonte: https://www.youtube.com/watch?v=RtA1rjhuavs)

Já o *`modelo com Hiperparametros`* notamos que ele não tem este problema, visto que a diferença entre as métricas de treino e teste são pequenas. Portanto, este modelo de arvore realmente aprendeu com os dados, este modelo por outro lado ao contrário do KNN e do Decision Tree é viável utilizar este modelo, ele realmente se saiu melhor em comparação ao modelo sem hiperparametros. 

**Ps: O `modelo com Hiperparametros` neste caso é realmente a melhor escolha, visto que, ele não sofre de overfitting, e métrica de acurácia é parecida. Este modelo se saiu melhor que o Decision Tree, KNN e Dummy.**

Mas embora já tenhamos um modelo *quase* campeão vamos estudar o gráfico de auc-roc para verificarmos melhor.

```python
# Importando a biblioteca para calcular 'Specificity' e 'Sensibility'
from sklearn.metrics import roc_curve
```

```python
# Retornando scores de cada modelo de machine learning treinada
y_prob_knn = knn_pipeline_best.predict_proba(X_test)[::,1]
y_prob_tree = tree_pipeline_best.predict_proba(X_test)[::,1]
y_prob_rf = rf_pipeline_best.predict_proba(X_test)[::,1]
y_prob_xg = xg_pipeline_best.predict_proba(X_test)[::,1]
```

```python
# Tirando dos scores fpr = 1 - Taxa de Falsos Positivos (Especificidade)
# tpr = 1 - Taxa de Verdadeiros Positivos (Sensibilidade)
# Valores que compoem 
fpr1, tpr1, _ = roc_curve(y_test, y_prob_knn)
fpr2, tpr2, _ = roc_curve(y_test, y_prob_tree)
fpr3, tpr3, _ = roc_curve(y_test, y_prob_rf)
fpr4, tpr4, _ = roc_curve(y_test, y_prob_xg)
```

```python
# Resultado das acuracia dos testes
auc1 = roc_auc_score(y_test, y_prob_knn).round(4)
auc2 = roc_auc_score(y_test, y_prob_tree).round(4)
auc3 = roc_auc_score(y_test, y_prob_rf).round(4)
auc4 = roc_auc_score(y_test, y_prob_xg).round(4)
```

```python
# Plotando a curva AUC-ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr1,tpr1,label="KNN, auc="+str(auc1))
plt.plot(fpr2,tpr2,label="Decision Tree, auc="+str(auc2))
plt.plot(fpr3,tpr3,label="Random Forest, auc="+str(auc3))
plt.plot(fpr4,tpr4,label="XGBoost, auc="+str(auc4))
plt.legend(loc=4, fontsize=12)
plt.ylabel('Sensibility')
plt.xlabel('Specificity')
plt.title('Gráfico AUC-ROC dos modelos de ML')
plt.show()
```

**Sobre a figura**

Conforme dito na introdução, os modelos KNN, Decision Tree estão praticamente fora de disputa.

O modelo de Random Forest e XGBoost já disputam qual melhor modelo, neste gráfico de curva AUC-ROC nota-se que o XGBoost tem melhor desempenho, mas pouco em relação ao Random Forest que acompanha logo em seguida.
