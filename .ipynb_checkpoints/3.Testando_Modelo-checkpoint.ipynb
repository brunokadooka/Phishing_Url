## 2. Criando as Features

```python
# Funcao para a Feature validação da URL, caso receba resposta no grupo de 200 é uma URL válida
def verifica_valid_url(url):
    url = "https://" + url if not url.startswith(('http', 'https')) else url
    try:
        response = requests.head(url, allow_redirects=True, timeout=1)  # Timeout definido como 1 segundo
        return 1 if response.status_code == 200 else 0
    except (requests.exceptions.Timeout, requests.exceptions.ConnectionError):
        return 0
    except requests.exceptions.MissingSchema:
        return 0
    except Exception:
        return 0
```

```python
# Função para obter as features de modelo para o dataset atual
def preparando_dataset(df):
    # Comprimento da URL
    df['url_length'] = df['URL'].apply(len)
    
    # Verificação se a URL é válida (usando um exemplo de função verifica_valid_url)
    df['valid_url'] = df['URL'].apply(verifica_valid_url)
    
    # Comprimento do caminho da URL
    df['path_length'] = df['URL'].apply(lambda x: len(urlparse(x).path))
    
    # Verifica se a URL possui o protocolo HTTPS
    df['isHttps'] = df['URL'].apply(lambda x: 1 if urlparse(x).scheme == 'https' else 0)
    
    # Contagem de pontos (.)
    df['nb_dots'] = df['URL'].apply(lambda x: x.count('.'))
    
    # Contagem de hifens (-)
    df['nb_hyphens'] = df['URL'].apply(lambda x: x.count('-'))
    
    # Verifica se a URL contém 'www'
    df['have_www'] = df['URL'].apply(lambda x: 1 if 'www' in x else 0)
    
    # Variável target
    df['target'] = (df['Label'] == 'good').astype(int)
    
    return df
```

```python
from joblib import Parallel, delayed

# Defina a função para obter as features de modelo
def preparando_features(row, idx):
    # Comprimento da URL
    row['url_length'] = len(row['URL'])
    
    # Verificação se a URL é válida
    row['valid_url'] = verifica_valid_url(row['URL'])
    
    # Comprimento do caminho da URL
    row['path_length'] = len(urlparse(row['URL']).path)
    
    # Verifica se a URL possui o protocolo HTTPS
    row['isHttps'] = 1 if urlparse(row['URL']).scheme == 'https' else 0
    
    # Contagem de pontos (.)
    row['nb_dots'] = row['URL'].count('.')
    
    # Contagem de hifens (-)
    row['nb_hyphens'] = row['URL'].count('-')
    
    # Verifica se a URL contém 'www'
    row['have_www'] = 1 if 'www' in row['URL'] else 0
    
    # Variável target
    row['target'] = 1 if row['Label'] == 'good' else 0
    
    print(f"Processando linha {idx}")  # Printa o índice da linha sendo processada
    return row

# Função para obter as features de modelo para o dataset atual de forma paralela
def preparando_dataset_paralelo(df):
    num_cores = 10
    
    # Usando Parallel para executar a função preparando_features em paralelo
    df_processed = Parallel(n_jobs=num_cores)(delayed(preparando_features)(row, idx) for idx, row in df.iterrows())
    
    return pd.DataFrame(df_processed)
```

```python
%%time
# Verificando tempo de execução da primeira função
df_test = preparando_dataset(df[:100])
```

```python
%%time
# Verificando tempo de execução da função com paralelismo
df_test = preparando_dataset_paralelo(df[:1000])
```

**Comentários**

Na função **`preparando_dataset(df)`**, criamos um dataset com as features necessárias para o modelo. No entanto, como observado na célula anterior, para um dataset de 100 linhas, a execução demora 57.2 segundos. A função mais custosa é **`verifica_valid_url(url)`**, que espera uma resposta da URL. Ajustamos o parâmetro de *timeout* para 1 segundo, aguardando assim um máximo de 1 segundo para obter uma resposta do site. Consideramos este tempo adequado, dado o custo computacional da função.

Apesar disso, o tempo de execução ainda é considerável, especialmente considerando os 549.346 registros. Para mitigar isso, podemos empregar a paralelização usando a biblioteca jolib, similar ao que fizemos para salvar os modelos no notebook de Modelagem. As funções que utilizamos, **`preparando_features(row, idx)`** e **`preparando_dataset_paralelo(df)`**, estão documentadas em [joblib.Parallel](https://joblib.readthedocs.io/en/latest/generated/joblib.Parallel.html), com exemplos semelhantes. Especificamente, o último exemplo demonstra como integrar funções com paralelismo.

Entretanto, precisamos de mais estratégias, pois mesmo para 1000 registros, leva-se 1 minuto. Considerando bruscamente que em 1 hora podemos processar 60 mil registros (assumindo 1 minuto para 1 mil registros), para os 549 mil registros precisamos de uma abordagem mais eficiente. Ao realizar os cálculos, notamos que 549.346 é divisível por 14, resultando em 39.239 registros por parte. Portanto, vamos adotar a estratégia de dividir o dataset em 14 partes, cada uma preparada e salva individualmente. Assim, podemos executar cada parte em um tempo razoável e, ao final, reunir todas as partes.


```python
# Funcao para dividir e salvar os datasets
def salvando_dataset_por_partes (df, idx_inicial, idx_final, part):
    
    print(f'Realizando a preparação da parte {part} - idx_inicio: {idx_inicial} e idx_final: {idx_final}....\n')
    
    # Inicio do tempo de execução
    inicio = time.time()
    
    # Repartindo o dataset
    df_test = df[idx_inicial:idx_final]
    
    # Realizando a preaparação das features
    df_test = preparando_dataset_paralelo(df_test)
    
    # Nome do arquivo
    name = f'data/dataset_parts/phishing_site_others_part{part}.pkl'
    df_test.to_pickle(name)
    
    # Fim do tempo de execução
    fim = time.time()
    
    # Calcular o tempo total
    tempo_total = fim - inicio  
    
    print(f'Concluido a parte {part} no tempo de {tempo_total} segundos....')
    print('----------------------------------------------------------')
```

```python
# Quantidade de partes
total_partes = 14

# Indice de partes
parte_atual = 1

# Laço para efetuar a divisao e realizar a tarefa
while (parte_atual <= total_partes):
    if (parte_atual == 1):
        idx_inicial = 0
        idx_final = int((parte_atual / total_partes) * len(df))
        salvando_dataset_por_partes (df, idx_inicial, idx_final, parte_atual)

    else:
        idx_inicial = idx_final
        idx_final = int((parte_atual / total_partes) * len(df))
        salvando_dataset_por_partes (df, idx_inicial, idx_final, parte_atual)
        
    parte_atual = parte_atual + 1
```

Salvando a imagem da execução

 


**Comentários**

A execução levou quase um dia, embora não fosse preciso deixar o tempo todo executando, poderiamos deixar para executar uma parte e executar outra, mas de qualquer forma temos as 14 partes de 39.239 registros. Assim, esta foi uma forma de lidar com muitos registros quando o PC leva um tempo enorme de execução, podemos acelerar o processo com a paralelização (Threads) e quebrar o problema em partes.

Agora vamos fazer um código que junte todas as partes e salvar este arquivo.

```python
# Criando um dataframe vazio
df_test = pd.DataFrame()

# Juntando as partes
for i in range(1, 15):
    
    # Nome do arquivo com as partes
    name = f'data/dataset_parts/phishing_site_others_part{i}.pkl'
    
    # df para importar o dataset
    df_import = pd.read_pickle(name)
    
    # Juntando os arquivos
    df_test = pd.concat([df_test, df_import])
    

# Salvando o dataset original (depois de juntar todas as partes)
df_test.to_pickle('data/dataset_parts/phishing_site_others_complete.pkl')
```

De qualquer forma como dito esta foi uma forma de resolver a questão de um dataset muito grande, podemos usar Threads e quebrar o dataset em partes e depois juntar tudo.

```python
# Importando a biblioteca para calcular 'Specificity' e 'Sensibility'
from sklearn.metrics import roc_curve

# Retornando scores de cada modelo de machine learning treinada
y_prob_knn = model_knn.predict_proba(X)[::,1]
y_prob_tree = model_tree.predict_proba(X)[::,1]
y_prob_rf = model_rf.predict_proba(X)[::,1]
y_prob_xg = model_xg.predict_proba(X)[::,1]

# Tirando dos scores fpr = 1 - Taxa de Falsos Positivos (Especificidade)
# tpr = 1 - Taxa de Verdadeiros Positivos (Sensibilidade)
# Valores que compoem 
fpr1, tpr1, _ = roc_curve(df_test['target'], y_prob_knn)
fpr2, tpr2, _ = roc_curve(df_test['target'], y_prob_tree)
fpr3, tpr3, _ = roc_curve(df_test['target'], y_prob_rf)
fpr4, tpr4, _ = roc_curve(df_test['target'], y_prob_xg)

# Resultado das acuracia dos testes
auc1 = roc_auc_score(df_test['target'], y_prob_knn).round(4)
auc2 = roc_auc_score(df_test['target'], y_prob_tree).round(4)
auc3 = roc_auc_score(df_test['target'], y_prob_rf).round(4)
auc4 = roc_auc_score(df_test['target'], y_prob_xg).round(4)
```

```python
# Plotando a curva AUC-ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr1,tpr1,label="KNN, auc="+str(auc1))
plt.plot(fpr2,tpr2,label="Decision Tree, auc="+str(auc2))
plt.plot(fpr3,tpr3,label="Random Forest, auc="+str(auc3))
plt.plot(fpr4,tpr4,label="XGBoost, auc="+str(auc4))
plt.legend(loc=4, fontsize=12)
plt.ylabel('Sensibility')
plt.xlabel('Specificity')
plt.title('Gráfico Curva AUC-ROC dos modelos de ML')
plt.show()
```

As explicações do porque os modelos não se saíram bem são muitas:

Lembrando, nosso dataset de treino tinha um pouco mais de mil linhas, e a validação ou teste para os modelos tinha um pouco mais de 300 linhas. Para um dataset com quase 550 mil registros é obvio que o modelo treinado não se sairia muito bem. Logo, como já falado desde o começo, sabiamos que o correto é deixar o maior dataset para treino e o menor para teste, esta situação serviu apenas como estudo.

**Segundo** Além do treinamento insuficiente, acredito que as features também não contribuem muito, 7 features para prever um comportamento como já visto na visão geral, algumas features tem pouco a oferecer na previsão, logo, seria necessário features com mais relevância. Infelizmente no dataset que treinamos já tinha pré-definido algumas features, mas neste ultimo temos a URL e podemos fazer o processo de *feature engineering* para criar features mais relevantes.
